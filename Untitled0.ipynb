### Draw Autocorrelation graph


```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
import statsmodels.api as sm
from pandas import read_csv
from pandas import datetime
from matplotlib import pyplot
import matplotlib.pyplot as plt
from pandas.plotting import autocorrelation_plot
def parsr(x):
    return datetime.strptime('190'+x,'%Y-%m')

series = read_csv('Salesx_.csv', header=0, parse_dates=[0],index_col=0,squeeze=True,date_parser=parsr)
print(series.head())

autocorrelation_plot(series)
pyplot.show()
```

### Split a Dataset into Train and Test Sets

```python
X = series.values
size = int(len(X) * 0.66)
train, test = X[0:size], X[size:len(X)]
history = [x for x in train]
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
import statsmodels.api as sm
fig = pyplot.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig =sm.graphics.tsa.plot_acf(series,lags=17,ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(series,lags=17,ax=ax2)
```

#### Calculate rmse

```python
X = series.values
size = int(len(X))
train, test = X[0:size], X[0:size]
history = [x for x in train]
print('Observations: %d' % (len(X)))
print('Training Observations: %d' % (len(train)))
print('Testing Observations: %d' % (len(test)))
```

```python
predictions = list()
# walk-forward validation
for t in range(len(test)):
  model = ARIMA(history, order=(6, 2, 1)) 
  model_fit = model.fit()
  output = model_fit.forecast()
  yhat = output[0] 
  predictions.append(yhat)
  obs = test[t]
  history.append(obs)
  print(t,'predicted=%f, expected=%f' % (yhat, obs))
# evaluate forecasts
rmse = sqrt(mean_squared_error(test, predictions))
print('-------------- Evaluation --------------')
print('Test RMSE: %.3f' % rmse)
```

### Draw the graphs for prediction

```python
pyplot.plot(test)
pyplot.plot(predictions, color='red') 
pyplot.show()
```
